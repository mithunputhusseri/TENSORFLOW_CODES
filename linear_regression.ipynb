```python
import numpy as mp
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
```

```python
!wget https://raw.githubusercontent.com/lazyprogrammer/machine_learning_examples/master/tf2.0/moore.csv
```

```python
data = pd.read_csv('moore.csv',header=None, on_bad_lines='skip').values
```

```python
x=data[:,0].reshape(-1,1)
y=data[:,1]
```

```python
plt.scatter(x,y)
```

```python
y = mp.log(y)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Input(shape=[1,]),tf.keras.layers.Dense(1)])
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(0.001,0.0),loss='mse')
```

```python
r = model.fit(x,y,epochs=1000)
```

```python
plt.plot(r.history['loss'])
```
